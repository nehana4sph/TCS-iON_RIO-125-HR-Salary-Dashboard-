```python
df1 = pd.read_csv('/content/HR_comma_sep_working.csv')
df1.head(5)
```

```python
df1.info()
```

```python
df1['salary'].replace(['high', 'medium', 'low'],
                        [0, 1, 2], inplace=True)
```

```python
df1.drop(['Department', 'left'], axis=1, inplace=True)
```

```python
df1.head(5)
```

```python
X = df1.drop('salary', axis=1) # we will predict salary category
y = df1['salary']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# evaluate multinomial logistic regression model
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
```

```python
# define the multinomial logistic regression model
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
```

```python
# define the model evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate the model and collect the scores
n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
# report the model performance
print('Mean Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
# fit the model on the whole dataset
model.fit(X, y)
```

```python
# define a single row of input data
row=[0.56,0.86,2,200,2,0,0]
```

```python
# predict the class label
salary_output=model.predict([row])
if salary_output == 0:
  print("Salary is Low")
elif salary_output==1:
  print("Salary is Medium")
else :
  print("Salary is High")
```
